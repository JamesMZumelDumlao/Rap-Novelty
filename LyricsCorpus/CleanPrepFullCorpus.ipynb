```python
''' PART 2: Merging corpus and census '''
corpus = pd.read_csv('./FullWorkingCorpus.csv')
census = pd.read_csv('../../RapperCensus/CleanedRapperCensus.csv')
```

```python
corpus.head()
census = census.drop(axis=1, columns=['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8'])
census = census.rename(columns = {'title':'Artist_name'})
census.head()
```

```python
df = pd.merge(corpus, census, how='outer', on='Artist_name', left_on=None, right_on=None,
         left_index=False, right_index=False, sort=False,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
```

```python
df = df.drop(axis = 1, columns = ['Unnamed: 0', 'number', 'pageid'])
```

```python
df.head()
df = df[df.Artist_name != 'PR'] # Removing all artists from Puerto Rico
```

```python
df.to_csv("FullDataset.csv", header = True, index = False)
```
