```r
year_count <- cities %>% select( count, Year) %>%group_by(Year)  %>% summarize(rappers = sum(count) )
```

```r
year_count
```

```r
# fig(10,10)
# library(ggthemes)
ggplot( data = year_count, 
       mapping = aes(
    x = Year, 
    y = rappers, label = Year)) + 
geom_point(shape=21, color="black", fill="#69b3a2", size=6) + 
     geom_segment(aes(
                    xend=c(tail(Year, n=-1), NA), 
                    yend=c(tail(rappers, n=-1), NA)
                  ), arrow=arrow(length=unit(0.3,"cm"))
      )  + theme_bw()
```

```r

# fig(20,20)
ggmap(bw_map) +
  geom_point(data = cities,
             aes(x = lon, y = lat, size = count),
             color = "red", alpha = 0.3)+
  scale_size_continuous(range = c(2, 9)) +
  labs(title = "test",
       size = "",
       color = NULL) +
  guides(color = guide_legend(override.aes = list(size = 6))) + theme(legend.position = "None") + 
facet_wrap(~ Year )

```
